```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)
```

La función de densidad de probabilidad (FDP) de la distribución uniforme esta dada por:

$$
f(x) = \begin{cases}
\frac{1}{b-a} & \text{for } a \le x \le b \\
0 & \text{otherwise}
\end{cases}
$$

Donde:
- $x$ es la variable
- $a$ es el valor mínimo
- $b$ es el valor máximo

```python
x = np.linspace(-2, 2, 100)
y = stats.uniform.pdf(x, loc=0, scale=1)

plt.plot(x, y)
plt.show()
```

La función de distribución acumulada (FDA) de la distribución uniforme esta dada por:

$$
F(x) = \begin{cases}
0 & \text{for } x < a \\
\frac{x-a}{b-a} & \text{for } a \le x \le b \\
1 & \text{for } x > b
\end{cases}
$$

Where:
- $x$ es la variable
- $a$ es el valor mínimo
- $b$ es el valor máximo

```python
x = np.linspace(-2, 2, 100)
y = stats.uniform.cdf(x, loc=0, scale=1)

plt.plot(x, y)
plt.show()
```

```python
uniform_samples = stats.uniform.rvs(loc=0, scale=10, size=1000)  # Uniform distribution
print("Uniform Distribution:")
print(f"Mean: {np.mean(uniform_samples):.4f}")
print(f"Range: [{np.min(uniform_samples):.4f}, {np.max(uniform_samples):.4f}]\n")
```

La función de distribución de probabilidad (FPD) de la distribución normal:

$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2}
$$

Where:
- $x$ es la variable
- $\mu$ es la media
- $\sigma$ es la desviación estándar

```python
x = np.linspace(-3, 3, 100)
y = stats.norm.pdf(x, loc=0, scale=1)

plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(-3, 3, 100)
y = stats.norm.pdf(x, loc=0, scale=1)
lower_bound = -0.6
upper_bound = 1.2
x_fill = x[(x >= lower_bound) & (x <= upper_bound)]
plt.fill_between(x_fill, stats.norm.pdf(x_fill, loc=0, scale=1), color='skyblue', alpha=0.5)
plt.plot(x, y)
plt.show()
```

la función de distribución acumulada de la distribución normal esta dada por:

$$
F(x) = \frac{1}{2}\left(1+\text{erf}\left(\frac{x-\mu}{\sigma\sqrt{2}}\right)\right)
$$

La función de error, denotada por erf(x), es una función especial que aparece en probabilidad, estadística, y ecuaciones parciales. Se define como:

$$ \text{erf}(x) = \frac{2}{\sqrt{\pi}} \int_0^x e^{-t^2} dt $$

En el contexto de la distribución normal acumulada, el término $\text{erf}\left(\frac{x-\mu}{\sigma\sqrt{2}}\right)$
 utiliza una variable estandarizada $\frac{x-\mu}{\sigma}$ y escalada por $\frac{1}{\sqrt{2}}$.


```python
x = np.linspace(-3, 3, 100)
y = stats.norm.cdf(x, loc=0, scale=1)

plt.plot(x, y)
plt.show()
```

La función de masa de probabilidad (FMP) de la distribución de Poisson está dada por:

$$
P(X=k) = \frac{\lambda^k e^{-\lambda}}{k!}
$$

Donde:
- $k$ es el número de ocurrencias (un entero no negativo)
- $\lambda$ es la tasa promedio de ocurrencias por intervalo (siempre $\lambda > 0$)
- $e$ es la base del logaritmo natural ($e \approx 2.71828$)
- $k!$ es el factorial de $k$

```python
x = np.arange(0, 10, 1)

y = stats.poisson.pmf(x, mu=2)

plt.plot(x, y)
plt.show()
```

La función de distribución acumulada (FDA) de la distribución de Poisson está dada por:

$$
F(k; \lambda) = P(X \le k) = \sum_{i=0}^{\lfloor k \rfloor} \frac{\lambda^i e^{-\lambda}}{i!}
$$

Donde:
- $k$ es el número de ocurrencias (un entero no negativo)
- $\lambda$ es la tasa promedio de ocurrencias por intervalo (siempre $\lambda > 0$)
- $e$ es la base del logaritmo natural ($e \approx 2.71828$)
- $i!$ es el factorial de $i$
- $\lfloor k \rfloor$ es la parte entera de $k$, ya que la distribución de Poisson es discreta.

```python
x = np.arange(0, 10, 1)

y = stats.poisson.cdf(x, mu=2)

plt.plot(x, y)
plt.show()
```

```python
v = stats.poisson.rvs(mu=5, size=1000)
plt.hist(v, bins=range(0, 15), density=True)
plt.show()
```

La funciòn de distribucion de probabilidad (FDP) de la distritución t esta dada por:

$$
f(t) = \frac{\Gamma\left(\frac{\nu+1}{2}\right)}{\sqrt{\nu\pi}\Gamma\left(\frac{\nu}{2}\right)}\left(1+\frac{t^2}{\nu}\right)^{-\frac{\nu+1}{2}}
$$

Where:
- $t$ es la variable
- $\nu$ grados de libertad
- $\Gamma$ functión Gamma

```python
x = np.linspace(-4, 4, 100)
y = stats.t.pdf(x,loc=0,scale=1,df=10)

plt.plot(x, y)
plt.show()
```

la función de distribución acumulada de la distribución t esta dada por:

$$
F(t; \nu) = I_{\frac{t^2}{\nu+t^2}}\left(\frac{1}{2}, \frac{\nu}{2}\right)
$$

Where:
- $t$ es la variable
- $\nu$ grados de libertad
- $I_x(a, b)$ es la función beta regularizada incompleta.

```python
x = np.linspace(-4, 4, 100)
y = stats.t.cdf(x,loc=0,scale=1,df=10)

plt.plot(x, y)
plt.show()
```

La función de densidad de probabilidad (FDP) de la distribución Gamma está dada por:

$$
f(x; \alpha, \beta) = \frac{1}{\Gamma(\alpha) \beta^\alpha}x^{\alpha-1}e^{-x/\beta} \quad \text{for } x > 0, \alpha > 0, \beta > 0
$$

Donde:
- $x$ es la variable
- $\alpha$ es el parámetro de forma (shape parameter)
- $\beta$ es el parámetro de escala (scale parameter)
- $e$ es la base del logaritmo natural
- $\Gamma(\alpha)$ es la función Gamma evaluada en $\alpha$

```python
x = np.linspace(0, 100, 100)
y = stats.gamma.pdf(x,a=2,scale=10)

plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(0, 100, 100)
y = stats.gamma.cdf(x,a=2,scale=10)

plt.plot(x, y)
plt.show()
```

# Ajuste de distribuciones de probabilidad

```python
# Generate mystery data (actually from a gamma distribution)

mystery_data = stats.gamma.rvs(a=2, loc=1.5, scale=2, size=1000)

plt.hist(mystery_data, bins=50, density=True, alpha=0.7, color='gray', label='Data')
plt.show()
```

```python

# Try fitting different distributions
distributions = [
    ('Normal', stats.norm),
    ('Gamma', stats.gamma),
    ('Log-normal', stats.lognorm),
    ('Exponential', stats.expon)
]

# Fit each distribution and compute goodness of fit
results = []
for name, dist in distributions:
    # Fit distribution
    params = dist.fit(mystery_data)

    # Perform Kolmogorov-Smirnov test
    ks_stat, p_value = stats.kstest(mystery_data, dist.name, args=params)

    print(f"\n{name} distribution:")
    print(f"Parameters: {params}")
    print(f"KS statistic: {ks_stat:.4f}")
    print(f"P-value: {p_value:.4f}")
```

```python
# Plot the mystery data and fitted distributions
plt.figure(figsize=(12, 6))

# Plot histogram of mystery data
plt.hist(mystery_data, bins=50, density=True, alpha=0.7, color='gray', label='Data')

# Plot fitted distributions
x = np.linspace(min(mystery_data), max(mystery_data), 100)
for name, dist in distributions:
    params = dist.fit(mystery_data)
    plt.plot(x, dist.pdf(x, *params), label=f'Fitted {name}', alpha=0.7)

plt.title('Mystery Data with Fitted Distributions', fontsize=20)
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend(fontsize=16)
plt.grid(True, alpha=0.3)
plt.show()
```

Distribución de Pareto de dos parametros

La función de densidad de probabilidad (FDP) de la distribución de Pareto de dos parámetros está dada por:

$$
f(x; \alpha, \beta) = \frac{\alpha \beta^\alpha}{x^{\alpha+1}} \quad \text{for } x \ge \beta, \alpha > 0
$$

Donde:
- $x$ es la variable
- $\alpha$ es el parámetro de forma (shape parameter), $\alpha > 0$
- $\beta$ es el parámetro de escala (scale parameter), $\beta > 0$. También representa el valor mínimo posible de $x$.

```python
from scipy.stats import pareto

# Example parameters
alpha = 2.5  # Shape parameter
xm = 10      # Scale parameter (minimum value)

# Generate 1000 random samples
samples = stats.pareto.rvs(b=alpha, loc=0, scale=xm, size=1000)
```

```python
import scipy.stats as stats
import numpy as np
x = np.linspace(xm, max(samples) * 1.1, 50) # Create a range of x values
pdf_values = stats.pareto.pdf(x, b=alpha, loc=0, scale=xm)
```

```python
cdf_values = pareto.cdf(x, b=alpha, loc=0, scale=xm)
```

```python
import numpy as np
x = np.linspace(xm, max(samples) * 1.1, 50) # Create a range of x values
cdf_values = pareto.cdf(x, b=alpha, loc=0, scale=xm)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
#plt.hist(samples, bins=50, density=True, alpha=0.6, label='Sample Histogram')
plt.plot(x, pdf_values, 'r-', lw=2, label='Pareto PDF')
plt.title('Two-Parameter Pareto Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```
