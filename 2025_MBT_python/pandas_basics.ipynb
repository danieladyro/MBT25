```python
import pandas as pd
from math import log
import numpy as np


s = pd.Series([1, 1, 1, 1, 2, 2, 2, 3, 3, 4])
s
```

```python
s.value_counts()  # Frecuencias absolutas
s.value_counts(normalize=True)  # Frecuencias relativas
s.min()  # Mínimo
```

```python
s.max()  # Máximo
s.mean()  # Media
```

```python
s.var()  # Varianza
s.std()  # Desviación típica
s.describe()  # Resume descriptivo
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Reload the original dataset
df = pd.read_csv('https://raw.githubusercontent.com/asalber/manual-python/master/datos/colesterol.csv', sep=',', decimal=',')

# Convert 'colesterol' to numeric, coercing errors
df['colesterol'] = pd.to_numeric(df['colesterol'], errors='coerce')

# Drop rows with NaN in 'colesterol' for visualization
df_cleaned = df.dropna(subset=['colesterol'])

plt.figure(figsize=(8, 6))
sns.histplot(df_cleaned['colesterol'], kde=True)
plt.title('Distribution of Cholesterol Levels')
plt.xlabel('Cholesterol')
plt.ylabel('Frequency')
plt.show()
```
